# **Prediction using Supervised Machine Learning**

 **Importing the libraries required**

```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
```

Importing the Data

Hence, we observe that there is a positive relationship between the number of hours and scores obtained

```
data.isnull().sum()
```

we will divide our data into attributes and Labels

```
x = data.iloc[:, :-1].values # x represents the array of values in "Hours" column of the data
y = data.iloc[:, 1].values   # y represents the array of values in "Score" column of the data
```

splitting the data into train and test sets using train_test_split() method of Scikit-Learn Library

```
# we have chosen 20% of the data for testing purpose and remaining 80% for training purpose
train_x,values_x, train_y, values_y = train_test_split(x,y,random_state =1)
```

```
model = LinearRegression()  # training the model

model.fit(train_x,train_y)
```

```
print(values_x) # testing the data
y = model.predict(values_x) # Predicting the Scores
```

```
# Comparing the actual values to our predicted values
df=pd.DataFrame({"Actual":  values_y,"predict": y})
df
```

```
print("Mean Absolute Error:", mean_absolute_error(values_y, y))
```

**what will be predicted score if a student studies for 9.25 hrs/ day?**

```
hours =9.25
own_pred = model.predict([[hours]])
print("No of Hours = {}".format(hours))
print("Predicted Score = {}".format(own_pred[0]))
```
